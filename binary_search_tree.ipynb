```python
class Node:
    
    def __init__(self, value):
        self.left = None
        self.right = None
        self.value = value
        

class BinarySearchTree:
    
    def __init__(self):
        self.root = None
    
    def find(value):
        node = self.root
        while node and node.value != value:
            node = node.left if node.value > value else node.right
        return node
        
    def find_min():
        node = self.root
        while node:
            node = node.left
        return node
    
    def find_max():
        node = self.root
        while node:
            node = node.right
        return node
    
    def insert(value):
        parent = None
        node = self.root
        while node:
            parent = node
            node = node.left if value < node.value else node.right
        insert_node = Node()
        if not parent:
            self.root = insert_node
            return
        if value < parent.value:
            parent.left = insert_node
        else:
            parent.right = insert_node
            
    def delete(value):
        parent = None
        node = self.root
        while node and node.value != value:
            parent = node
            node = node.left if node.value > value else node.right
        if not node: return
        
        # 该节点有左右子节点
        if node.left and node.right:
            replace_node = node.right
            replace_node_parent = node
            while replace_node.left:
                replace_node_parent = replace_node
                replace_node = replace_node.left
            node.value = replace_node.value  # 替换值
            parent, node = replace_node_parent, replace_node  #  改为删除替换节点
        
        # 该节点为叶子节点或只有一个子节点
        child = node.left if node.left else node.right
        if not parent:
            self.root = child
        elif parent.left == node:
            parent.left = child
        else:
            parent.right = child
```
